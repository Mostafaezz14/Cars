```python
import numpy as np
import pandas as pd
df = pd.read_csv('Car_Models.csv')
```

```python
df.head()
```

```python
df.columns.str.strip()
```

```python
df.info()
```

```python
df['Price'] = df['Price'].str.replace('$', '').str.replace(',' , '.')
df['Horsepower'] = df['Horsepower'].str.replace('hp', '')
df['Torque'] = df['Torque'].str.replace('lb-ft', '')
```

```python
df.head()
```

```python
df['Horsepower'] = pd.to_numeric(df['Horsepower'] , errors = 'coerce')
df['Torque'] = pd.to_numeric(df['Torque'] , errors = 'coerce')
df['Number of Doors'] = pd.to_numeric(df['Number of Doors'] , errors = 'coerce').astype('Int64')
df['Price'] = pd.to_numeric(df['Price'] , errors = 'coerce')
df['Number of Cylinders'] = pd.to_numeric(df['Number of Cylinders'] , errors = 'coerce').astype('Int64')
```

```python
df.info()
```

```python
df['Company'] = df['Company'].str.lower()
df['Model'] = df['Model'].str.lower()
df['Body Type'] = df['Body Type'].str.lower()
```

```python
df
```

```python
df.isnull().sum()
```

```python
df['Horsepower'] = df['Horsepower'].fillna(0)
df['Torque'] = df['Torque'].fillna(0)
df['Fuel Economy'] = df['Fuel Economy'].fillna('unknow')
df['Number of Doors'] = df['Number of Doors'].fillna(0)
df['Price'] = df['Price'].fillna(0)
df['Model Year Range'] = df['Model Year Range'].fillna('unknow')
df['Body Type'] = df['Body Type'].fillna('unknow')
df['Engine Type'] = df['Engine Type'].fillna('unknow')
df['Number of Cylinders'] = df['Number of Cylinders'].fillna(0)
```

```python
df.isnull().sum()
```

```python
df.drop_duplicates(keep=False, inplace=True)
```

```python
df.duplicated().any()
```

# Q1. highest model have hoursepower

```python
df[df['Horsepower'].max() == df['Horsepower']][['Horsepower' , 'Model']]
```

# Q2. how many models in bmw company

```python
df[df["Company"].str.contains("bmw" , case = False)]["Company"].count()
```

# Q3. how many coupe cars 

```python
df[df["Body Type"].str.contains("coupe" , case = False)]["Body Type"].count()
```

# Q4. highest 5 models have Torque

```python
df.sort_values(by=['Torque'], ascending = False).head(5)
```

# Q5. highest 5 cars horsepower in 2023-2024

```python
df[df["Model Year Range"].str.contains("2023-2024" , case = False)].sort_values(by=['Horsepower'], ascending = False)[['Company','Model','Horsepower']].head(5)
```

# Q6. Company and the model have electric engine

```python
df[df["Engine Type"].str.contains("Electric" , case = False)].sort_values(by=['Company'], ascending = False)[['Company','Model','Engine Type']]
```
